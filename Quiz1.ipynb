```python
secuence=''
for l in lines[1:]: # muestra la iteración
  secuence+=l[:-1] # secuencia en una sola líena
print(secuence) 
```

Ejemplo Bioinformático de busqueda por metodo de fuerza bruta


```python
problema='AATGAATTCTCGTAACTACATAGCACAAGTAGA'# subcadena a encontrar
a=secuence.find(problema) #inicio de la secuencia llamada problema
print(a)
```

```python
#SOLUCION

genoma = secuence
print(len(genoma)) # cuenta los elementos de la variable subcadena
print(genoma.find('AATGAATTCTCGTAACTACATAGCACAAGTAGA')) # nos dice donde inicia nuestra subcadena
subcadena='AGTAGA'#fragmento final minima de nuestra subcadena que no se repite en el resto de la cadena
print(genoma.rfind(subcadena)) # inicio de la subcadena
```
